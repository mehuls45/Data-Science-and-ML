```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

# Distribution plots

```python
# Distribution plots - Helps to understand the distribution of data
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.distplot(tips['total_bill'],kde=False) # kde = Kernel Density Estimator
```

```python
# Joint plot - Kind of scatter plot
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips) 
```

```python
# Scatter plot with regression line
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
# To show density
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
# Pairplot - Scatter plot for every numerice value with other
```

```python
sns.pairplot(tips)
```

```python
# We can additionaly add hue to the data
```

```python
sns.pairplot(tips,hue='sex')
```

```python
# kdeplot - Shows on Kernel Density Estimator
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical plots

```python
# categorical plots helps to understand the relation between numerical and categorical data
```

```python
# Barplot - More like groupby
```

```python
sns.barplot(x='sex',y='tip',data=tips)
```

```python
# We can even count the occurences of attribute value using countplot
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
# Boxplot - To show distribution of data
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
# Violinplot - To show distribution of data
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
```

## Matrix plots 

### i. Heat Map

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
# Heat Maps require that the data should be in matrix form of correlation table, so we make a correlation table
```

```python
tc = tips.corr()
```

```python
tc
```

```python
sns.heatmap(tc)
```

```python
sns.heatmap(tc,annot=True)
```

```python
sns.heatmap(tc,cmap='coolwarm')
```

```python
sns.heatmap(tc,cmap='coolwarm',annot=True)
```

```python
# We will now plot a heat map for the flight table
```

```python
fp = flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
fp
```

```python
sns.heatmap(fp,cmap='magma')
```

```python
# Linecolor and Linewidths can be customized to show clear seperation
```

```python
sns.heatmap(fp,cmap='magma',linecolor='white',linewidths=1)
```

### ii. Cluster Map

```python
# Cluster map groups similar kind of data so that it becomes more visually useful
```

```python
sns.clustermap(fp)
```

```python
# For better visualizton we can standard the data we are plotting
```

```python
sns.clustermap(fp,standard_scale=1)
```

```python
# Now, we have much more clearer picture about the data
```

```python
# lmplot - Linear Model Plot
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# We can conclude that, male and female both have the same fit line
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
# Instead of color, we can even seperate based on different attributes
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```
